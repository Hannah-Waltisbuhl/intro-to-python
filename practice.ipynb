```python
import numpy as np
import copy

def create_puzzle_array(puzzleString):
    """
    """
    return np.array(list(puzzleString)).reshape(9,9).astype(int)

def print_puzzle_array(puzzleArray):
    """
    """
    for row in range(9):
        for col in range(9):
            print(puzzleArray[row][col], end=" ")
            if col % 3 == 2:
                print(" ", end = "")
        print("\n", end = "")
        if row % 3 == 2:
            print("\n", end = "")


def get_row(puzzleArray, rowIndex):
    """
    """
    return puzzleArray[rowIndex]

def get_col(puzzleArray, colIndex):
    """
    """
    transposedArray = puzzleArray.transpose()
    return transposedArray[colIndex]

def get_square(puzzleArray, rowIndex, colIndex):
    """
    """
    startRow = rowIndex // 3 * 3
    startCol = colIndex // 3 * 3
    squareArray = puzzleArray[startRow:startRow + 3, startCol:startCol + 3]
    return squareArray

def get_possible_numbers(puzzleArray, rowIndex, colIndex):
    """
    """
    row = get_row(puzzleArray, rowIndex)
    col = get_col(puzzleArray, colIndex)
    square = get_square(puzzleArray, rowIndex, colIndex)
    possibleNumbers = []

    if puzzleArray[rowIndex][colIndex] != 0:
        possibleNumbers.append(puzzleArray[rowIndex][colIndex])
        return possibleNumbers

    for digit in range(1,10):
        if digit not in row and not digit in col and digit not in square:
            possibleNumbers.append(digit)

    return possibleNumbers

def get_next_cell(rowIndex, colIndex):
    """
    """
    nextCol = (colIndex + 1) % 9
    nextRow = rowIndex
    finished = False
    if nextCol == 0:
        nextRow = (rowIndex + 1) % 9
        if nextRow == 0 and nextCol == 0:
            finished = True
    
    return finished, nextRow, nextCol


def solve_sudoku(puzzleArray, rowIndex, colIndex):
    """
    """
    tempPuzzle = copy.deepcopy(puzzleArray)

    while True:
        possibleNumbers = get_possible_numbers(tempPuzzle, rowIndex, colIndex)

        if not possibleNumbers:
            return False
        
        finished, nextRow, nextCol = get_next_cell(rowIndex, colIndex)

        if finished:
            tempPuzzle[rowIndex][colIndex] = possibleNumbers[0]
            return tempPuzzle

        if len(possibleNumbers) > 1:
            break
        
        tempPuzzle[rowIndex][colIndex] = possibleNumbers[0]
        rowIndex = nextRow
        colIndex = nextCol

    for number in possibleNumbers:
        tempPuzzle[rowIndex][colIndex] = number
        result = solve_sudoku(tempPuzzle, nextRow, nextCol)

        if type(result) != type(False):
            return result
        
    return False


    

    
```
